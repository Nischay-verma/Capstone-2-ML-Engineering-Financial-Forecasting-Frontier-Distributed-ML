###**1:  Data Loading and Basic Inspection**

#### Load the bank.csv dataset into a Spark DataFrame.
#### Perform basic data inspection tasks such as showing the first few rows, printing the schema, and displaying a summary of numerical columns.


```python
#ques 1 Load the bank.csv dataset into a Spark DataFrame.
#Perform basic data inspection tasks such as showing the first few rows, printing the schema, and displaying a summary of numerical columns.

# Step 1: Setup SparkSession
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Bank Data Analysis") \
    .getOrCreate()

# Step 2: Load the dataset
df = spark.read.csv('/content/bank (1).csv', header=True, inferSchema=True)

# Step 3: Show first few rows
df.show(5)

# Step 4: Print schema
df.printSchema()

# Step 5: Summary of numerical columns
df.describe().show()
```

The commands df.show(5), df.printSchema(), and df.describe().show() are used for initial data exploration. df.show(5) displays the first five records of the dataset, df.printSchema() shows the structure and data types of all columns, and df.describe().show() provides summary statistics such as count, mean, standard deviation, minimum, and maximum values for numerical features. Together, these steps help verify that the dataset has been loaded correctly and provide a basic understanding of the data distribution.

###**2:  Data Filtering and Column Operations**


#### Filter the dataset to include only clients with a balance greater than a certain threshold (e.g., 1000).
#### Extract the month from the 'month' column and create a new column for the quarter of the year.


```python
# ques 2 Filter the dataset to include only clients with a balance greater than a certain threshold (e.g., 1000).
#Extract the month from the 'month' column and create a new column for the quarter of the year.

from pyspark.sql.functions import col, when

# Step 1: Filter clients with balance > 1000
filtered_df = df.filter(col("balance") > 1000)
filtered_df.show(5)

# Step 2: Create quarter column based on 'month'
# Mapping month to quarter
quarter_df = filtered_df.withColumn("quarter", when(col("month").isin("jan", "feb", "mar"), "Q1")
                                     .when(col("month").isin("apr", "may", "jun"), "Q2")
                                     .when(col("month").isin("jul", "aug", "sep"), "Q3")
                                     .when(col("month").isin("oct", "nov", "dec"), "Q4")
                                     .otherwise("Unknown"))
quarter_df.select("month", "quarter").distinct().show()


```

The filtering operation successfully extracted clients with a balance greater than 1000, allowing focused analysis of higher-value customers. Additionally, a new quarter column was created by mapping months to their respective quarters (Q1–Q4). This transformation enables easier quarterly trend analysis and helps identify seasonal patterns in customer behavior and campaign performance.

###**3:  GroupBy and Aggregation**

#### Group the data by job type and calculate the average balance and median age for each job category.
#### Find the total number of clients for each marital status who have subscribed to a term deposit.


```python
# ques 3 Group the data by job type and calculate the average balance and median age for each job category.
#Find the total number of clients for each marital status who have subscribed to a term deposit.

from pyspark.sql.functions import avg, count, expr, percentile_approx

# Step 1: Average balance and median age by job type
job_stats = quarter_df.groupBy("job").agg(
    avg("balance").alias("avg_balance"),
    percentile_approx("age", 0.5).alias("median_age")
)
job_stats.show(truncate=False)

# Step 2: Total number of clients per marital status who subscribed to term deposit
marital_subscribed = quarter_df.filter(col("y") == "yes") \
    .groupBy("marital") \
    .agg(count("*").alias("subscribed_clients"))
marital_subscribed.show()


```

The analysis shows that retired and housemaid clients have the highest average account balances, while student clients have the lowest median age, reflecting expected demographic and financial patterns. These findings provide valuable insights into the financial characteristics of different job categories. Additionally, married clients recorded the highest number of term deposit subscriptions, suggesting that marital status may influence investment decisions. Such insights can help banks design more targeted marketing campaigns and customer engagement strategies.

###**4:  Use UDF to Categorize Age Groups**

#### Write a UDF that categorizes clients into age groups ('<30', '30-60', '>60') and apply it to create a new column age_group.

```python
# ques 4 Write a UDF that categorizes clients into age groups (<30, 30-60, >60) and apply it to create a new column age_group.

from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Step 1: Define the UDF function
def age_group(age):
    if age < 30:
        return "<30"
    elif age <= 60:
        return "30-60"
    else:
        return ">60"

# Step 2: Register the UDF
age_group_udf = udf(age_group, StringType())

# Step 3: Apply the UDF to create new column
age_grouped_df = quarter_df.withColumn("age_group", age_group_udf(col("age")))
age_grouped_df.select("age", "age_group").distinct().orderBy("age").show(10)
```

The new age_group column correctly assigns clients under 30 to the <30 category.

This will help in segmentation and targeted analysis by age category.



###**5:  Advanced Data Transformations**

#### Calculate the subscription rate (percentage of clients who subscribed to a term deposit) for each education level.
#### Identify the top 3 professions that have the highest loan default rate.


```python
# ques 5 Calculate the subscription rate (percentage of clients who subscribed to a term deposit) for each education level.

#Identify the top 3 professions that have the highest loan default rate.

from pyspark.sql.functions import round

# 1. Subscription Rate by Education Level
edu_total = age_grouped_df.groupBy("education").agg(count("*").alias("total_clients"))
edu_subscribed = age_grouped_df.filter(col("y") == "yes").groupBy("education").agg(count("*").alias("subscribed"))

# Join to calculate subscription rate
edu_subscription_rate = edu_total.join(edu_subscribed, on="education", how="left") \
    .fillna(0) \
    .withColumn("subscription_rate", round((col("subscribed") / col("total_clients")) * 100, 2))
edu_subscription_rate.select("education", "subscription_rate").show()

# 2. Top 3 Job Types with Highest Loan Default Rate
loan_total = age_grouped_df.groupBy("job").agg(count("*").alias("total_clients"))
loan_yes = age_grouped_df.filter(col("loan") == "yes").groupBy("job").agg(count("*").alias("loan_yes"))

loan_default_rate = loan_total.join(loan_yes, on="job", how="left") \
    .fillna(0) \
    .withColumn("loan_default_rate", round((col("loan_yes") / col("total_clients")) * 100, 2))

top3_loan_defaults = loan_default_rate.orderBy(col("loan_default_rate").desc()).limit(3)
top3_loan_defaults.select("job", "loan_default_rate").show()


```

The subscription rate by education level measures the percentage of clients within each education category who subscribed to a term deposit (y = yes). This analysis helps identify the education groups that are most responsive to marketing campaigns and financial products.

The analysis of loan defaults by profession highlights the top job categories with the highest proportion of clients having loans. This provides insights into financial obligations across different occupations and can assist banks in assessing risk profiles and developing targeted lending strategies.

###**6:  String Manipulation and Date Functions**

#### Concatenate the 'job' and 'marital' columns into a new column 'job_marital'.
#### Use string functions to convert the 'contact' column to uppercase.


```python
# ques 6 Concatenate the 'job' and 'marital' columns into a new column 'job_marital'.

#Use string functions to convert the 'contact' column to uppercase.

from pyspark.sql.functions import concat_ws, upper

# 1. Concatenate 'job' and 'marital' into 'job_marital'
concat_df = age_grouped_df.withColumn("job_marital", concat_ws("_", col("job"), col("marital")))

# 2. Convert 'contact' column to uppercase
string_df = concat_df.withColumn("contact_upper", upper(col("contact")))

# Display the new columns
string_df.select("job", "marital", "job_marital", "contact", "contact_upper").show(5)

```

Two new features were created to enhance the dataset. The job_marital column combines a client's job and marital status into a single attribute (e.g., technician_single), which can capture interactions between these characteristics. The contact_upper column standardizes the contact method by converting all values to uppercase (e.g., cellular to CELLULAR). These transformations improve data consistency and can be useful for reporting, analysis, and machine learning models.

###**8:  Complex Queries for Insights**

#### Analyze which month of the year has the highest number of clients contacted and determine the success rate of the campaign in that month.

```python
# ques 8 Analyze which month of the year has the highest number of clients contacted,
#and determine the success rate of the campaign in that month.

# Step 1: Count of clients contacted in each month
monthly_counts = string_df.groupBy("month").agg(count("*").alias("total_contacts"))

# Step 2: Count of successful subscriptions in each month
monthly_success = string_df.filter(col("y") == "yes") \
    .groupBy("month").agg(count("*").alias("successful_contacts"))

# Step 3: Join and calculate success rate
monthly_summary = monthly_counts.join(monthly_success, on="month", how="left") \
    .fillna(0) \
    .withColumn("success_rate", round((col("successful_contacts") / col("total_contacts")) * 100, 2)) \
    .orderBy(col("total_contacts").desc())

monthly_summary.show()

```

The table summarizes campaign performance by month, showing the total number of client contacts, successful subscriptions, and the corresponding success rate. This analysis helps identify the most active campaign periods and the months with the highest conversion rates, providing valuable insights for optimizing future marketing and outreach strategies.

####  Find the average duration of contact for clients who subscribed (y = yes) versus those who did not (y = no)

```python
# ques 9 Find the average duration of contact for clients who subscribed (y = yes) versus those who did not (y = no).

from pyspark.sql.functions import mean

# Group by 'y' (yes/no) and calculate average contact duration
avg_duration_by_subscription = string_df.groupBy("y").agg(round(mean("duration"), 2).alias("avg_duration"))
avg_duration_by_subscription.show()
```

The analysis shows that clients who subscribed to a term deposit had an average contact duration of approximately 484 seconds, compared to about 233 seconds for clients who did not subscribe. This suggests a positive relationship between longer customer interactions and successful subscriptions, indicating that extended conversations may improve the likelihood of conversion.

###**11:  Contact Method Analysis**

#### Investigate which contact method ('cellular', 'telephone', etc.) has the highest success rate for term deposit subscriptions.


```python
# ques 12 Investigate which contact method ('cellular', 'telephone', etc.) has the highest success rate for term deposit subscriptions.
# Total contacts per contact method
contact_total = string_df.groupBy("contact").agg(count("*").alias("total_contacts"))

# Successful subscriptions per contact method
contact_success = string_df.filter(col("y") == "yes") \
    .groupBy("contact").agg(count("*").alias("successful_contacts"))

# Join and calculate success rate
contact_analysis = contact_total.join(contact_success, on="contact", how="left") \
    .fillna(0) \
    .withColumn("success_rate", round((col("successful_contacts") / col("total_contacts")) * 100, 2)) \
    .orderBy(col("success_rate").desc())

contact_analysis.show()

```

The analysis compares the effectiveness of different contact methods, such as cellular, telephone, and unknown. For each method, it reports the total number of contacts, the number of successful term deposit subscriptions, and the corresponding success rate. These insights help identify which communication channel is most effective in converting clients and can guide future marketing and outreach strategies.

###**12:  Data Processing with Spark SQL**

#### Calculate the average account balance for different age groups or the most common job types. (Use CREATE TEMPORARY VIEW in Spark SQL to query data using SQL syntax.)

```python
# ques 13 Calculate the average account balance for different age groups or the most common job types.
#(Use CREATE TEMPORARY VIEW in Spark SQL to query data using SQL syntax.)

# Register the Spark DataFrame as a SQL temporary view
string_df.createOrReplaceTempView("bank_data")

# 1. Average account balance for each age group
spark.sql("""
    SELECT age_group, ROUND(AVG(balance), 2) AS avg_balance
    FROM bank_data
    GROUP BY age_group
    ORDER BY age_group
""").show()

# 2. Top 5 most common job types
spark.sql("""
    SELECT job, COUNT(*) AS job_count
    FROM bank_data
    GROUP BY job
    ORDER BY job_count DESC
    LIMIT 5
""").show()


```

The analysis indicates that clients aged over 60 generally maintain higher account balances, while clients under 30 tend to have lower balances. This insight can help banks design age-specific financial products and marketing strategies. Additionally, the majority of customers belong to the management and blue-collar job categories, highlighting these groups as key customer segments for targeted campaigns and service offerings.
